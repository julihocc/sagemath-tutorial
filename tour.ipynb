It's straightforward to use Sage to perfom linear algebra operations. For example, you can create matrices and vectors, and perform operations like addition, subtraction, multiplication, and inversion. ALthough, this could accomplished with the `numpy`, `scipy`, `sympy` and related library in Python, Sage provides a more user-friendly interface, with a more complete set of functionalities.

```python
A = matrix([[1,2],[3,4]])
B = A^(-1) 
A*B 
```

Also, you can use Sage as as CAS (Computer Algebra System) to solve equations, simplify expressions, and plot functions.

```python
a = var('a')
integrate(sqrt(a)*sqrt(1+a), a).simplify()
```

```python
x = var('x')
S = solve(x^2 - 2*x - 8 == 0, x)    
S[0].rhs(), S[1].rhs()
```

```python
show(plot(sin(x) + sin(1.6*x), 0, 40))
```
